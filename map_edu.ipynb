```python
# 將 'geometry' 列重新命名為 'county_geometry'
County_data = County_data.rename(columns={'geometry': 'county_geometry'})

# 合併 'Town_data' 和 'County_data'
merged_data = Town_data.merge(County_data[['COUNTYNAME', 'county_geometry']], on='COUNTYNAME', how='left')

# 將合併後的 'county_geometry' 列的值插入 'Town_data' 的 'geometry' 列
merged_data['geometry'] = merged_data['county_geometry']

# 刪除多餘的 'county_geometry' 列
merged_data.drop(['county_geometry'], axis=1, inplace=True)

# 顯示合併後的 GeoDataFrame
merged_data
```

```python
merged_data['geometry']
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 將資料轉換為 GeoDataFrame，並將 'geometry' 列重新命名為 'geom'
gdf = gpd.GeoDataFrame(merged_data, geometry='geometry')
gdf.rename_geometry('geom', inplace=True)

# 畫出地圖
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='population', cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# 添加標題
plt.title('Population Distribution', fontdict={'fontsize': '15', 'fontweight' : '3'})

# 顯示圖表
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 將 'population' 列的資料型別轉換為整數
gdf['population'] = gdf['population'].astype(int)

# 將人口數除以 50
gdf['population_scaled'] = gdf['population'] // 50

# 畫出地圖
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='population_scaled', cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# 添加標題
plt.title('Population Distribution', fontdict={'fontsize': '15', 'fontweight' : '3'})

# 顯示圖表
plt.show()
```
