```python
import pandas as pd
import geopandas as gpd

# Read CSV file
Rice_data_csv = pd.read_csv("COA_OpenData.csv")

# Convert to GeoDataFrame
Rice_data = gpd.GeoDataFrame(Rice_data_csv)


Rice_data
```

```python
import pandas as pd

# 假設 Rice_data 是你的 DataFrame
# Rice_data = pd.DataFrame(...)

# 以 city_area 列為基準，將 yield 列加總
result = Rice_data.groupby('city_area')['yield'].sum().reset_index()

# 顯示結果
result
```

```python
import pandas as pd

# 假設 result 是你的 DataFrame
# 假設 County_data 是你的 DataFrame

# 將 result 和 County_data 合併
merged_data = pd.merge(result, County_data, left_on='city_area', right_on='COUNTYNAME', how='left')
# 刪除含有 NaN 值的列
cleaned_data = merged_data.dropna()
# 保留需要的欄位
cleaned_data = cleaned_data[['yield', 'COUNTYNAME', 'geometry']]

# 重新設定索引
cleaned_data.reset_index(drop=True, inplace=True)

# 顯示合併後的結果
cleaned_data
```

```python
import numpy as np
merged_data[['yield']] = merged_data[['yield']].astype('int64')

# 檢查yield列的類型
print(merged_data['yield'].dtype)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 將資料轉換為 GeoDataFrame，並將 'geometry' 列重新命名為 'geom'
gdf = gpd.GeoDataFrame(merged_data, geometry='geometry')
gdf.rename_geometry('geom', inplace=True)

# 畫出地圖
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='yield', cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# 添加標題
plt.title('yield', fontdict={'fontsize': '15', 'fontweight' : '3'})

# 顯示圖表
plt.show()
```
